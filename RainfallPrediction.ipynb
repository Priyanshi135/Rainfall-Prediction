3. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv("austin_weather_final.csv")
```

```python
data
```

```python
x=data.drop(["PrecipitationSumInches"],axis=1)
y=data["PrecipitationSumInches"]
```

4. Reshaping it into 2D vector

```python
y=y.values.reshape(-1,1)
y
```

```python
days_index=786
days=[i for i in range(y.size)]
```

5. Initializing the linear regression classifier

```python
clf=LinearRegression()
clf.fit(x,y)
```

6. Plotting a graph

```python
print("The precipitation trend graph: ")
plt.scatter(days,y,color='g')
plt.scatter(days[days_index],y[days_index],color='r')
plt.title("Precipitation level")
plt.xlabel("days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis=x.filter(['TempAvgF','DewPointAvgF','HumidityAvgPercent','SeaLevelPressureAvgInches','VisibilityAvgMiles','WindAvgMPH'])
```

```python
print("The precipitation vs attributes trend graph: ")
for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i + 1)
    plt.scatter(days, x_vis[x_vis.columns.values[i]], color='g')
    plt.scatter(days[days_index], x_vis.iloc[days_index, i], color='r')
    plt.title(x_vis.columns.values[i])
plt.show()
```
